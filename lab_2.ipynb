```python
import numpy as np
import pandas as pd
heroes_information = pd.read_csv("heroes_information.csv")
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
encabezados = """
    SELECT *  FROM heroes_information LIMIT 100;
"""
pysqldf(encabezados)
```

```python
#Cuál es el nombre del primer superhéroe de la tabla heroes?
query_1 = """
        SELECT name FROM heroes_information where id = 0;
"""
pysqldf(query_1)
```

```python
#Cuántas casas publicadoras existen?
query_2 = """
        SELECT COUNT(DISTINCT Publisher) FROM heroes_information ;
"""
pysqldf(query_2)
```

```python
#Cuántos superhéroes miden más de 2 metros (200 cms)?
query_3 = """
        SELECT COUNT(*) FROM heroes_information WHERE Height > 200;
"""
pysqldf(query_3)
```

```python
#Cuántos superhéroes son humanos y miden más de 2 metros (200 cms)?
query_4 = """
        SELECT COUNT(*) FROM heroes_information WHERE Height > 200 AND upper(RACE) = upper('Human');
"""
pysqldf(query_4)
```

```python
#Cuántos superhéroes pesan más de 100 lbs y menos de 200 lbs?
query_5 = """
        SELECT COUNT(*) FROM heroes_information WHERE Weight BETWEEN 100 AND 200;
"""
pysqldf(query_5)
```

```python
#Cuántos superhéroes tienen los ojos de color azul o rojo?
query_6 = """
        SELECT COUNT(*) FROM heroes_information WHERE `Eye color` = 'blue' or `Eye color` = 'red';
"""
pysqldf(query_6)
```

```python
#Cuántos superhéroes son Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro
query_7 = """
        SELECT COUNT(*) FROM heroes_information 
        WHERE (Race in ('Human', 'Mutant') AND `Hair color` = 'Green') 
        OR (Race = 'Vampare' AND `Hair color` = 'Black');
"""
pysqldf(query_7)
```

```python
#Cuál es el nombre del primer superhéroe si ordeno la tabla por raza en orden descendente?
query_8 = """
        SELECT name FROM heroes_information ORDER BY Race DESC;
"""
pysqldf(query_8)
```

```python
#Cuántos superhéroes son de género masculino y cuántas de género femenino?
query_9 = """
        SELECT Gender, COUNT(*) FROM heroes_information group by gender;
"""
pysqldf(query_9)
```

```python
#Cuántas casas publicadoras tienen más de 15 superhéroes?
query_10 = """
    SELECT Publisher, count(*) FROM heroes_information group by Publisher HAVING count(*) >= 15;
    
"""
pysqldf(query_10)
```
